```python


class MorphologicalFSA:
    def __init__(self):
        self.prefixes = {
            "un": "NEGATION",
            "re": "REPETITION",
            "dis": "NEGATION"
        }

        self.suffixes = {
            "ing": "PROGRESSIVE",
            "ed": "PAST",
            "s": "PLURAL",
            "es": "PLURAL",
            "ness": "NOUN",
            "ly": "ADVERB"
        }

    def analyze(self, word):
        state = "q0"
        original_word = word
        word = word.lower()

        prefix = None
        suffix = None
        root = word
        features = []


        for p in self.prefixes:
            if word.startswith(p) and len(word) > len(p) + 2:
                state = "q1"
                prefix = p
                root = word[len(p):]
                features.append(self.prefixes[p])
                break
        if state == "q0":
            state = "q2"
        else:
            state = "q2"
        for s in sorted(self.suffixes, key=len, reverse=True):
            if root.endswith(s) and len(root) > len(s) + 1:
                state = "q3"
                suffix = s
                root = root[:-len(s)]
                features.append(self.suffixes[s])
                break
        state = "q4"
        return {
            "word": original_word,
            "prefix": prefix,
            "root": root,
            "suffix": suffix,
            "features": features,
            "final_state": state
        }
sample_words = [
    "cats",
    "walked",
    "running",
    "unhappy",
    "replayed",
    "happiness",
    "quickly",
    "disliked"
]
fsa = MorphologicalFSA()

print("==== Finite State Automata Morphological Analysis ====\n")

for word in sample_words:
    result = fsa.analyze(word)
    print(result)
```
