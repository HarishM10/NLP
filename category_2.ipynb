```python
import math
from collections import Counter
import random
corpus = [
    "I love NLP",
    "I love machine learning",
    "NLP is fun",
    "machine learning is powerful"
]
def preprocess(sentences):
    words = []
    for sentence in sentences:
        tokens = sentence.lower().split()
        words.extend(tokens)
    return words
words = preprocess(corpus)
random.shuffle(words)
split_index = int(0.75 * len(words))

train_words = words[:split_index]
test_words = words[split_index:]
def build_unigram_model(words):
    total_words = len(words)
    counts = Counter(words)
    probabilities = {word: count / total_words for word, count in counts.items()}
    return probabilities, total_words
train_probs, total_train = build_unigram_model(train_words)
def compute_entropy(probabilities):
    entropy = 0
    for p in probabilities.values():
        entropy -= p * math.log2(p)
    return entropy
entropy = compute_entropy(train_probs)
def compute_cross_entropy(train_probs, test_words):
    epsilon = 1e-10
    total = len(test_words)
    cross_entropy = 0
    for word in test_words:
        prob = train_probs.get(word, epsilon)
        cross_entropy -= math.log2(prob)
    return cross_entropy / total
cross_entropy = compute_cross_entropy(train_probs, test_words)
def compute_perplexity(value):
    return 2 ** value
perplexity_train = compute_perplexity(entropy)
perplexity_test = compute_perplexity(cross_entropy)
print("Train Words:", train_words)
print("Test Words:", test_words)
print("--------------------------------")
print(f"Entropy (Train): {entropy:.4f}")
print(f"Perplexity (Train): {perplexity_train:.4f}")
print("--------------------------------")
print(f"Cross-Entropy (Test): {cross_entropy:.4f}")
print(f"Perplexity (Test): {perplexity_test:.4f}")
```
