```python
import nltk
import re
nltk.download('brown')
from nltk.corpus import brown
words = ['Talking','ended','singing','complete']
sample_words = list(set(words[:3]))
def morphological_analyzer(word):
    analysis = {
        "word": word,
        "prefix": None,
        "root": word,
        "suffix": None,
        "features": []
    }
    prefixes = {
        "un": "NEGATION",
        "re": "REPETITION",
        "in": "NEGATION",
        "dis": "NEGATION"
    }
    for prefix, meaning in prefixes.items():
        if word.startswith(prefix) and len(word) > len(prefix) + 2:
            analysis["prefix"] = prefix
            analysis["root"] = word[len(prefix):]
            analysis["features"].append(meaning)
            break
    root = analysis["root"]
    suffix_rules = [
        ("ing", "PROGRESSIVE"),
        ("ed", "PAST"),
        ("ly", "ADVERB"),
        ("ness", "NOUN"),
        ("ment", "NOUN"),
        ("s", "PLURAL"),
        ("es", "PLURAL")
    ]
    for suffix, feature in suffix_rules:
        if root.endswith(suffix) and len(root) > len(suffix) + 1:
            analysis["suffix"] = suffix
            analysis["root"] = root[:-len(suffix)]
            analysis["features"].append(feature)
            break
    return analysis
print("==== Rule-Based Morphological Analysis (Brown Corpus) ====\n")
for word in sample_words[:30]:
    result = morphological_analyzer(word)
    print(result)
```
