```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')


dataset = [
    "This is a note",
    "He is a student in this collage",
    "Running and singing is fun!!!",
    "Cristiano ROnaldo and messi is goat"
]



stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()



def preprocess_text(text):
    print("\nOriginal Text:", text)


    text = text.lower()
    print("After Lowercasing:", text)
    text = re.sub(r'[^a-z\s]', '', text)
    print("After Cleaning:", text)

    tokens = word_tokenize(text)
    print("After Tokenization:", tokens)

    tokens_no_stop = [word for word in tokens if word not in stop_words]
    print("After Stopword Removal:", tokens_no_stop)

    stemmed_words = [stemmer.stem(word) for word in tokens_no_stop]
    print("After Stemming:", stemmed_words)

    lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens_no_stop]
    print("After Lemmatization:", lemmatized_words)

    return {
        "tokens": tokens,
        "no_stopwords": tokens_no_stop,
        "stemmed": stemmed_words,
        "lemmatized": lemmatized_words
    }




print("====== TEXT PREPROCESSING PIPELINE ======")

for sentence in dataset:
    preprocess_text(sentence)
```
